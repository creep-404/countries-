To read, load, explore, and understand the Countries dataset using the Pandas library in Python


```python

import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("countries.csv")

# Display the first five rows of the dataset
print("First 5 Rows:")
print(df.head())

# Display the last five rows of the dataset
print("Last 5 Rows:")
print(df.tail())

# Shape of the dataset
print("Shape of the Dataset (Rows, Columns):", df.shape)

# Column names
print("Column Names:")
print(df.columns)

# Dataset info (Data types + Non-null counts)
print("Dataset Info:")
df.info()

# Summary statistics for numerical columns
print("Descriptive Statistics:")
print(df.describe())

# Data types of each column
print("Data Types:")
print(df.dtypes)

# Missing values in each column
print("Missing Values:")
print(df.isnull().sum())

# Unique values in each column
print("Unique Values:")
print(df.nunique())

# Display 5 random sample rows
print("5 Random Sample Rows:")
print(df.sample(5))

# Value counts for the Continent column
print("Value Counts for 'Continent' Column:")
print(df["Continent"].value_counts())







```
